# Data Preparation

```python
data = original_data.copy()
```

```python
data = data.iloc[:,1:]    # ไม่เอาid เพราะไม่ใช้
data
```

## Missing Value

```python
data.isna().sum()
```

```python
medianAge = data['Age'].median()
data['Age'] = data['Age'].fillna(medianAge)
# เติมค่าmedianAgeแทนค่าnull เพื่อลบค่าnullออกไป
```

```python
data.isna().sum()     # ไม่มีค่าnullที่ageแล้ว
```

```python
data = data.dropna()    # แถวไหนมีค่าnull ลบแถวนั้นออกเลย
```

```python
data.isna().sum()
```

## Categorical Data

```python
data['Gender'].value_counts()
```

```python
data['Gender'] = data['Gender'].replace(['M','F'],['Male','Female'])
```

```python
data['Gender'].value_counts()
```

```python
data['Driving_License'].value_counts()
```

```python
data = data[data['Driving_License'].isin([0,1])]  # เอาแต่ค่า0กับ1 dataที่rowไหนมีค่า-1เอาออกหมด
```

```python
data['Driving_License'].value_counts()
```

```python
data['Vehicle_Age'].value_counts()
```

```python
data['Vehicle_Age'] = data['Vehicle_Age'].replace(['< 1 Year','1-2 Year','> 2 Years'],[0,1,2])  #โมเดลชอบตัวเลข ไม่ชอบตัวอักษร
```

```python
data['Vehicle_Age'].value_counts()
```

```python
data['Vehicle_Damage'].value_counts()
```

```python
data['Vehicle_Damage'] = data['Vehicle_Damage'].replace(['Yes','No'],[1,0])  #โมเดลชอบตัวเลข ไม่ชอบตัวอักษร
```

```python
data['Vehicle_Damage'].value_counts()
```

```python
for column in data.columns:
  print(column)
  print(np.sort(data[column].unique()))
  print('------------')
```

## Change Data Type

```python
data["Response"].dtype
```

```python
data["Response"] = pd.to_numeric(data["Response"], downcast='integer')
```

```python
data["Response"]
```

```python
data["Age"] = pd.to_numeric(data["Age"], downcast='integer')
```

```python
data["Age"]
```

## Encode Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['Gender'].unique()
```

```python
GenderEncoder = LabelEncoder()
```

```python
GenderEncoder.fit(data['Gender'].unique())  # ทำให้modelรู้ว่า gender มีแค่2ตัว คือ 'Male' กับ 'Female'
```

```python
GenderEncoder.classes_
```

```python
GenderEncoder.transform(['Female', 'Male'])   # แปลงเป็นตัวเลข
```

```python
GenderEncoder.inverse_transform([1,0,0,1])
```

```python
data['Gender'] = GenderEncoder.transform(data['Gender'])
```

```python
data
```

```python
for column in data.columns:
  print(column)
  print(np.sort(data[column].unique()))
  print('------------')
```

## Outlier
[Empirical Rule](https://statisticsbyjim.com/probability/empirical-rule/)

```python
data
```

|data - mean| <= 3*SD

```python
data = data[np.abs(data['Annual_Premium']-
            data['Annual_Premium'].mean()) <= 
            (3*data['Annual_Premium'].std())]
data
```

```python
data.isna().sum()   
```

```python
data['Gender'].unique()
```

```python
data['Age'].unique()
```

```python
data['Spending Score (1-100)'].unique()
```

```python
data = data.dropna()
data.isna().sum()   
```

```python
data['Gender'] = data['Gender'].replace(['M','F'],['Male','Female'])
data['Gender'] = data['Gender'].replace(['Male','Female'],[1,0])
data['Gender'].unique()
```

```python
sns.boxplot(data=data['Age']) 
```

```python
sns.boxplot(data=data['Annual Income (k$)'])
```

```python
sns.boxplot(data=data['Spending Score (1-100)']) 
```

```python
data = data[np.abs(data['Age']-
            data['Age'].mean()) <= 
            (3*data['Age'].std())]
data = data[np.abs(data['Annual Income (k$)']-
            data['Annual Income (k$)'].mean()) <= 
            (3*data['Annual Income (k$)'].std())]
data = data[data['Annual Income (k$)'] >= 0]
data
```

```python
data["Age"] = pd.to_numeric(data["Age"], downcast='integer')
for column in data.columns:
  print(column)
  print(np.sort(data[column].unique()))
  print('------------')
```

```python
data = data.reset_index(drop=1)
data
```

```python
data.to_csv('mallcusto_cleaned.csv', index=False)
```
