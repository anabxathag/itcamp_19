### Make_Blob_Data

```python
from sklearn.datasets import make_blobs
```

```python
X,Y = make_blobs(n_samples=100, centers=3,n_features=2,random_state=0,cluster_std=.5)
#สร้างข้อมูล 100 ตัว , มี 2 ตัวเเปร, สร้างข้อมูล 3 กลุ่ม,  สุ่มได้ค่าเดิม, การกระจายข้อมูล 0.5 
```

```python
X
```

```python
X.shape
```

```python
Y #บอกว่าข้อมูลเเต่ละตัวอยู่กลุ่มไหน
```

```python
plt.plot(X[:,0],X[:,1],'o')
```

### Model Construction

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.get_params()
```

```python
kmeans.fit(X,);
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
Y #บอกว่าข้อมูลเเต่ละตัวอยู่กลุ่มไหน
```

```python
kmeans.inertia_ #E2 error
```

```python
kmeans.n_iter_ #ดูว่ามันวนทำซ้ำหาจุดcentroidไปกี่รอบ
```

### Model Deployment

```python
#Predict
kmeans.predict(X)
```

```python
X
```

```python
#Compue Cluster Distance (ข้อมูล x ห่างจาก กลุ่ม 0 1 2 เท่าไหร่)
kmeans.transform(X)
```

```python
#opposite sign of E2 error
kmeans.score(X)
```

```python
#Fit and Predict
kmeans.fit_predict(X)
```

```python
#Fit and Predict
kmeans.fit_transform(X)
```

### Find Number of Cluster

```python
K = range(2,15)
E2 =[]
for k in K:
    model=KMeans(n_clusters=k)
    model = model.fit(X)
    E2.append(model.inertia_)
E2
```

```python
plt.plot(K,E2,'rx-')
```

```python
plt.plot(X[:,0],X[:,1],'o')
```

```python
MX1 = kmeans.cluster_centers_[:,0]
MX2 = kmeans.cluster_centers_[:,1]
```

```python
print(MX1)
print(MX2)
```

```python
plt.plot(X[:,0],X[:,1],'o')
plt.plot(MX1,MX2,'+')
```
