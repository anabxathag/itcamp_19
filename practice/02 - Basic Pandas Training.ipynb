#Pandas
Pandas เป็นไลบรารี Python ที่ใช้สำหรับการจัดการและวิเคราะห์ข้อมูลที่มีโครงสร้าง โดยมีการนำเข้าข้อมูลเป็นตาราง (table) ที่เรียกว่า DataFrame ซึ่งจะมีลักษณะคล้ายกับตารางในโปรแกรม Excel ซึ่งมีหัวข้อและเเถวเเนวตั้งเป็นตัวบอกรายละเอียดของข้อมูลต่างๆ ด้วยฟังก์ชั่นที่มีอยู่ใน Pandas เราสามารถทำการดึงข้อมูลจากไฟล์ CSV, Excel, SQL, หรือแม้กระทั่งจากฐานข้อมูล NoSQL มาใช้งานได้ง่ายดาย นอกจากนี้ Pandas ยังมีความสามารถในการจัดการ Missing Values, การกรองข้อมูล (filtering), การรวมข้อมูล (merging), การจัดการข้อมูลซ้ำ (duplicates) และการวิเคราะห์ข้อมูลอื่นๆ ที่เป็นประโยชน์ในการวิเคราะห์และจัดการข้อมูลต่างๆ

### Concat Column of Datafame

### เฉลย

```python
order = pd.read_csv('/content/order.csv')
order
```

```python
payment = pd.read_csv('/content/payment.csv')
payment
```

```python
merged = order.join(payment.set_index('order_id'), on='order_id', rsuffix='_payment')
#rsuffix เป็น parameter ที่ใช้กำหนด suffix ให้กับคอลัมน์ที่ชื่อซ้ำกันในขณะที่ merge กัน ในที่นี้จะใช้ '_payment' 
#แทนส่วนที่ซ้ำกันของคอลัมน์ status ใน DataFrame payment จะถูกแทนด้วย 'status_payment'
```

```python
merged
```

```python
merged = merged.iloc[:,1:] #เลือกตั้งเเต่ 0 ถึง 1 ทั้ง row and col
merged
```

```python
merged['total_amount'] = merged['quantity'] * merged['price']
merged
```

```python
res = merged.loc[merged['status_payment']=='completed',['order_id','status_payment','total_amount']]
res
```
