```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/it19_data/Food Composition.csv')
df
```

```python
ans = df['Classification Name']
print(ans.unique().shape)
print(ans.unique())
```

```python
data = df[['Food Name', 'Food Description']]
data
```

```python
data.info()
```

```python
data.isna().sum()  
```

```python
for i in data.columns:
  print(df[i].unique().shape)
  print(df[i].unique())
  print()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
vectorizer.fit_transform(data['Food Name'])
vectorizer.get_feature_names_out()
```

```python
data_foodname = pd.DataFrame()
categ = vectorizer.get_feature_names_out()[18:]
for i in categ:
  ans_foodname = []
  for j in data['Food Name']:
    if i in j.lower():
      ans_foodname.append(1)
    else:
      ans_foodname.append(0)
  data_foodname[i] = ans_foodname
data_foodname
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit_transform(data['Food Description'])
vectorizer.get_feature_names_out()[:100]
```

```python
data_description = pd.DataFrame()
categ = vectorizer.get_feature_names_out()[30:]
for i in categ:
  ans_description = []
  for j in data['Food Description']:
    if i in j.lower():
      ans_description.append(1)
    else:
      ans_description.append(0)
  data_description[i] = ans_description
data_description
```

```python
datafood = pd.concat([data_foodname, data_description], axis=1)
datafood
```

```python
category, y = np.unique(ans, return_inverse=True)
y
```

# Supervised Learning

## Neural Network

```python
n_values = np.max(y) + 1
y = np.eye(n_values)[y]
y
```

```python
from sklearn.model_selection import train_test_split

train, test, train_Y, test_Y = train_test_split(datafood, y, test_size=0.2, random_state=28)
train.sample(10)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(2832,)))
model.add(Dense(512, activation='sigmoid'))
model.add(Dense(280, activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(train, train_Y, epochs=100, verbose=1, batch_size=32, validation_data=(test, test_Y))
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("The test accuracy score is", accuracy_score(test_Y.argmax(axis=1), model.predict(test).argmax(axis=1)))
```

# Unsupervised learning

```python
from sklearn.cluster import KMeans
```

```python
K = range(2,20)
E2 =[]
for k in K:
    model_3= KMeans(n_clusters=k)
    model_3 = model_3.fit(datafood)
    E2.append(model_3.inertia_)
E2
```

```python
plt.plot(K,E2,'rx-')
```

```python
model_3 = KMeans(n_clusters=3)
model_3.fit(datafood)
```
