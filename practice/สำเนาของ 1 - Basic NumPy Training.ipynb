# Adding and Removing Elements

## Adding Elements

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
np.concatenate((a, b))
```

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6]])
np.concatenate((x, y),axis=0)
```

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6]])
np.concatenate((x, y))
```

```python
x = np.array([[1, 2]])
y = np.array([[5, 6]])
np.concatenate((x, y),axis=1)
```

## Removing Elements


```python
arr08 = np.arange(1,21).reshape((4,5))
arr08
```

```python
np.delete(arr08, 1, axis=0)
```

```python
np.delete(arr08, [0,2,4], axis=1)
```

## Sum, Min, Max

```python
print(f'sum: {np.sum(arr09)}')
print(f'min: {np.min(arr09)}')
print(f'max: {np.max(arr09)}')
```

```python
np.min(arr09, axis=1)
```

```python
np.max(arr09, axis=0)
```

## Mean, Median, Standard Deviation, Variance

```python
print(f'mean: {np.mean(arr09)}')
print(f'median: {np.median(arr09)}')
print(f'SD: {np.std(arr09)}')
print(f'variance: {np.var(arr09)}')
```

```python
np.mean(arr09, axis=1)
```
